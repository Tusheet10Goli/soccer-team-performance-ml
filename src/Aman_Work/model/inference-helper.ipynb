```python
import pandas as pd
```

```python
players = pd.read_csv("player-data")
team = pd.read_csv("team-data")

# find statistically average team:
df = pd.read_csv("team-data")

home = df[[col for col in df.columns if (col == "Date" or col.startswith("Home"))]].copy(deep=True)
home.rename(mapper = lambda x: x.replace("Home", "Team"), inplace=True, axis=1)

away = df[[col for col in df.columns if (col == "Date" or col.startswith("Away"))]].copy(deep=True)
away.rename(mapper = lambda x: x.replace("Away", "Team"), inplace=True, axis=1)

data = pd.concat([home, away])
statsList = list(filter(lambda x: x.startswith("Team "), data.columns))

average = dict()
for stat in statsList:
    average[stat] = data[stat].mean()

averageTeam = pd.DataFrame([average])
averageTeam.rename(mapper = lambda x: x.replace("Team ", ""), inplace=True, axis=1)

for stat in averageTeam.columns:
    assert stat in players.columns
    
averageTeam
```

```python
averagePlayers = list()
for player in players["Player"].unique():
    tmpDict = dict()
    tmpDict["Player"] = player
    playerStats = players[players["Player"] == player]
    if len(playerStats) > 100:
        playerStats = playerStats.sort_values("Date").tail(100) # last 100 appearances
    
    for stat in averageTeam.columns:
        tmpDict[stat] = playerStats[stat].mean()
    
    tmpDict["Minutes"] = playerStats["Minutes"].mean()
    
    averagePlayers.append(tmpDict)
averagePlayers = pd.DataFrame(averagePlayers)
for stat in averageTeam.columns:
    assert stat in averagePlayers.columns
assert len(averagePlayers) == len(players["Player"].unique())
averagePlayers
```

```python
averageTeam.to_csv("average-team", index=False)
averagePlayers.to_csv("average-players", index=False)
```
